# 2. Use the five percent significance test over the data to determine the p value to check new scheme has significantly raised outputs? (10 points)

H0 : 𝜇old – 𝜇new = 0 (New Scheme did not raise the output
significantly)
H1 : 𝜇old – 𝜇new < 0 (New Scheme raised the output significantly)

It is asked that whether the new scheme has significantly raised the output, it is an example of the one-tailed t-test.

```python
old_data = titan_data['Old Scheme (in thousands)']
```

```python
new_data = titan_data['New Scheme (in thousands)']
```

```python
# check for the normality of old scheme data
from scipy.stats import shapiro, ttest_rel
from statsmodels.stats.power import ttest_power

t_statistic, p_value = shapiro(old_data)
print('t=%f, p=%f' % (t_statistic, p_value))

#p-value is > 0.05, data is normally distributed
```

```python
# check for the normality of new scheme data
t_statistic, p_value = shapiro(new_data)
print('t=%f, p=%f' % (t_statistic, p_value))

#p-value is > 0.05, data is normally distributed
```

```python
# since both old and new data are normally distributed, using paired ttest to determine if new scheme has raised outputs
t_statistic, p_value = ttest_rel(old_data, new_data)
print('t=%f, p=%f' % (t_statistic, p_value))

# the value returned here is a two tailed p-value. The one tailed p-value will be 0.130576/2=0.06528
```

# 3. What conclusion does the test (p-value) lead to? (2.5 points)

In this scenario, the p value is 0.06528 which is greater than the 0.05.
Hence accept the null hypothesis that the new scheme did NOT
raise the output significantly
